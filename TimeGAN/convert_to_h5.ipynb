```
import pandas as pd
import quandl
from pathlib import Path
import json

DEFAULT_OUTPUT_DIR = Path('./h5-output')
DEFAULT_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

hdf_store = DEFAULT_OUTPUT_DIR / 'WIKI_train.h5'
json_file = DEFAULT_OUTPUT_DIR / 'dataset.json'
```

```
quandl.ApiConfig.api_key = ""
start_date = '1990-01-01'
end_date = '2019-02-01'
```

```
tickers = ['AAPL', 'IBM', 'MSFT', 'WMT']
data = pd.DataFrame(columns=tickers)
for ticker in tickers:
    data[ticker] = quandl.get('WIKI/' + ticker, start_date=start_date,
    end_date=end_date)['Adj. Close']
data.head()
```

```
data.to_hdf(hdf_store, "table")
```

```
def dump_json(json_dict):

    if json_file.exists():
        with json_file.open('r') as f:
            ori_json_dict = json.load(f)
        ori_json_dict.update(json_dict)
    else:
        ori_json_dict = json_dict

    with json_file.open('w') as f:
        json.dump(ori_json_dict, f)
```

```
num_data = data.shape[0]
print(num_data)

json_dict = {'num_data': num_data}
dump_json(json_dict)
```
